#### Custom Floating Project

This tutorial uses the prepared ORBIT config file `~/configs/custom_floating_project.yaml`. This example uses `ParametricManager()` to sweep different turbine sizes to demonstrate the scaling abilities of the `CustomSemiSubmersibleDesign()` class.  

Last updated: July 2024

```python
import pandas as pd
from pprint import pprint
from ORBIT import load_config, ProjectManager, ParametricManager

weather = pd.read_csv("data/example_weather.csv", parse_dates=["datetime"])\
            .set_index("datetime")
```

### Load the project configuration
`~/configs/custom_floating_project.yaml`

```python
floating_config = load_config("configs/custom_floating_project.yaml")

print(f"Num turbines: {floating_config['plant']['num_turbines']}")
print(f"Turbine: {floating_config['turbine']}")
print(f"\nSite: {floating_config['site']}")
```

### Phases

```python
print("Design phases:\n")
pprint(floating_config['design_phases'])
print(f"\nInstall phases: \n")
pprint(list(floating_config['install_phases'].keys()))
```

### Run

```python
project = ProjectManager(floating_config, weather=weather)
project.run()
```

### Top Level Outputs

```python
print(f"Installation CapEx:  {project.installation_capex/1e6:.0f} M")
print(f"System CapEx:       {project.system_capex/1e6:.0f} M")
print(f"Turbine CapEx:       {project.turbine_capex/1e6:.0f} M")
print(f"Soft CapEx:          {project.soft_capex/1e6:.0f} M")
print(f"Total CapEx:        {project.total_capex/1e6:.0f} M")

print(f"\nInstallation Time: {project.installation_time:.0f} h")
```
